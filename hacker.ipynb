```python
import pandas as pd
from sodapy import Socrata
pathCard = "/Users/wendellliu/Downloads/Fare_Card_History_for_Metropolitan_Transportation_Authority__MTA___Beginning_2010.csv"
pathMTA = "/Users/wendellliu/Downloads/MTA_Subway_Hourly_Ridership__Beginning_February_2022.csv"
dfMTA = pd.read_csv(pathMTA)
```

```python
from datetime import datetime
# import requests
pathWeather = "/Users/wendellliu/Downloads/weather.csv"
df_w = pd.read_csv(pathWeather)
flag = 0
dfMTA['weather'] = "null"
for index, row in dfMTA.iterrows():
    date_str = row['transit_timestamp']
    stamp = datetime.strptime(date_str, '%m/%d/%Y %I:%M:%S %p').timestamp()
    weather_data_series = df_w[df_w['dt'] == stamp]['weather_main']
    
    # Check if we got any matching weather data, if so, assign the first value.
    if not weather_data_series.empty:
        dfMTA.at[index, 'weather'] = weather_data_series.iloc[0]

    flag += 1
    if flag % 500000 == 0:
        print(flag, dfMTA.at[index, 'weather'])

dfMTA.to_csv('/Users/wendellliu/Code/Python/MTA.csv', index=False)
dfMTA.head(2)



```
